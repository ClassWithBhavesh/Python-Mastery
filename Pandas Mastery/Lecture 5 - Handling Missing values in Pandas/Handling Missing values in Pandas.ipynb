### ***Deleting Columns in Pandas***




```python
import pandas as pd
```

```python
prod_df = pd.DataFrame(pd.read_csv("product_data.csv"))
prod_df

df = prod_df.head(15)
newDf = df.drop(columns='list_price', inplace=False)
newDf1 = df.drop(columns='list_price', inplace=False)

print(newDf)
print(newDf1)

# upDf = newDf + newDf1 # it will marge the categorized data

updated_df = pd.concat([newDf, newDf1])   # it will concat the provided arguments
updated_df.drop_duplicates(inplace=True)  # this method will remove the dup records (rows ya columns)
updated_df
```
