### ***Exception Handling in Python***

```python
# Errors - while writing or perform a program if an issue is encountered then it will be an Error.
# 3 tarike ki hoti hai errors -
  #  1. Compile-Time Errors - jo compile karte samay issue ya bugs aare usse complile time error kahenge
  #  2. Run-Time Error      - jo error run karte samay aaye
  #  3. Logical Errors      - jab expected outcomes naa aaye tab logical error hoga
```

```python
# Exception Handling to handle the run-time errors -
# using try-except-finally


# DocString - ''' or """
# code = """
#   if userInp == 0
#     print("Error - you can't divide anything with 0")
#   print(25/userInp, " - division")
# """
# try:
#   exec(code)
# except SyntaxError:
#   print("Error - you have to use ':' in conditional statement to follow the indentation")

try:
  userInp = int(input("Enter the number - "))
  print(f"division is - {25/userInp}")

except ZeroDivisionError:
  print("you can't divide anything with zero, try provide different value other than '0'")

except NameError:
  print("Error - you have not defined userInp")

except ValueError:
  print("Error - the input is not an integer, please provide integer value to perform the division operation")

finally:
  print("Exception handling successfully done!")
```
