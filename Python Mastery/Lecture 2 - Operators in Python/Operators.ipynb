```python
# Operators in Python -
# operators - symbol which will perform certain operations
# operands  - operands are the value on which we will perform operations
# operation - is a task or a process in which something is done

# Arithmatic
# Assignment
# Comparion
# Logical

# Arithmatic Operators : +, -, *, /, **, //, %
# num1 = 25
# num2 = 21
# print(num1 + num2)
# tSum = num1 + num2
# print("Sum of num1 and num2 : ", tSum)


num1 = 25
num2 = 21
tSum = num1 - num2
print("Sub of num1 and num2 : ", tSum)


num1 = 25
num2 = 21
tSum = num1 * num2
print("Multiplication of num1 and num2 : ", tSum)


num1 = 25
num2 = 5
tSum = num1 / num2          # it will perform division on the operands and return only Decimal Values
print("Division of num1 and num2 : ", tSum)


num1 = 25
num2 = 5
tSum = num1 // num2         # it will perform division on the operands and return only Integer Values
print("Floor Division of num1 and num2 : ", tSum)


num1 = 25
num2 = 2
tSum = num1 ** num2
print("Exponential of num1 over num2 : ", tSum)


num1 = 25
num2 = 26
tSum = num1 % num2            # Remainder of the division performed
print("Modulus of num1 and num2 : ", tSum)
```
